```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Carregando dataset
df = pd.read_csv('student_habits_performance.csv', delimiter=',')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='O')
```

```python
#Traduzindo para português brasileiro
df.columns = ['Id_do_estudante','Idade','Gênero','Horas de Estudo por Dia','Média de Horas em Redes Sociais',
              'Média de horas no Netflix','Emprego de meio período?','Porcentagem de Presença',
              'Horas de Sono','Qualidade de Dieta','Frequência de Exercícios',
              'Nível Educacional dos Pais','Qualidade da Internet','Avaliação de Saúde Mental',
              'Participação em Atividades Extracurriculares','Nota da Prova']
```

```python
df.head(10)
```

```python
df.info()
```

```python
#Tratamento de dados ausentes
moda = df["Nível Educacional dos Pais"].mode()[0]
df['Nível Educacional dos Pais'] = df['Nível Educacional dos Pais'].fillna(moda)
df.head(10)
```

```python
#Tratamento de dados - Traduzindo para o Português-BR
df['Gênero'] = df['Gênero'].map({'Male': 'Masculino','Female': 'Feminino','Other':'Outro'})
df['Emprego de meio período?'] = df['Emprego de meio período?'].map({'No':'Não','Yes':'Sim'})
df['Qualidade de Dieta'] = df['Qualidade de Dieta'].map({'Fair':'Regular','Good':'Boa','Poor':'Pobre'})
df['Nível Educacional dos Pais'] = df['Nível Educacional dos Pais'].map({'Bachelor':'Bacharel','High School':'Ensino Médio','Master':'Mestrado'})
df['Qualidade da Internet'] = df['Qualidade da Internet'].map({'Average':'Regular','Good':'Boa','Poor':'Pobre'})
df['Participação em Atividades Extracurriculares'] = df['Participação em Atividades Extracurriculares'].map({'No':'Não','Yes':'Sim'})
```

```python
df.head(10)
```

```python
f, ax = plt.subplots(2, 3, figsize=(15, 8))  # 2 linhas, 3 colunas
ax = ax.flatten()  # transforma matriz 2x3 em vetor de 6 elementos

sns.countplot(data=df, x='Gênero', ax=ax[0], color='Darkblue')
sns.countplot(data=df, x='Emprego de meio período?', ax=ax[1], color='Darkblue')
sns.countplot(data=df, x='Qualidade de Dieta', ax=ax[2], color='Darkblue')
sns.countplot(data=df, x='Nível Educacional dos Pais', ax=ax[3], color='Darkblue')
sns.countplot(data=df, x='Qualidade da Internet', ax=ax[4], color='Darkblue')
sns.countplot(data=df, x='Participação em Atividades Extracurriculares', ax=ax[5], color='Darkblue')

plt.tight_layout()
plt.show()
```

```python
graficos = df[['Idade','Horas de Estudo por Dia','Média de Horas em Redes Sociais',
               'Frequência de Exercícios',
               'Avaliação de Saúde Mental']]

for x in graficos:
    plt.figure(figsize=(10, 5))
    plt.hist(df[x], edgecolor='white', color='darkblue', bins=len(df[x].unique()))
    plt.title(f'Histograma de {x}')
    plt.xlabel(x)
    plt.ylabel('Quantidade')
    plt.tight_layout()
    plt.show()
```

```python
# Calcula a matriz de correlação
corr = df.corr(numeric_only=True) 

# Cria o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Mapa de Calor - Correlação entre Variáveis')
plt.tight_layout()
plt.show()
```

```python
#Setando variaveis correlacionadas
x = df['Horas de Estudo por Dia']
y = df['Nota da Prova']

#Coeficiente de correlação
coef = np.polyfit(x,y,1)
linha_de_tendencia = np.poly1d(coef)

#plotando o gráfico de dispersão
plt.figure(figsize=(10,6))
plt.scatter(x, y,   color='darkblue', alpha=0.5)
plt.plot(x,linha_de_tendencia(x), color='red', label='Linha de Tendência')
plt.legend()
plt.title('Relação entre Horas de Estudo por dia e a Nota da Prova')
plt.ylabel('Nota da Prova')
plt.xlabel('Horas de Estudo por Dia')
plt.show()
```
